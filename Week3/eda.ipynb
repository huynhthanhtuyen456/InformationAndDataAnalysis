```python
df.columns
```

```python
df = df.copy()
df = df[['coaster_name', 
        #  'Length', 'Speed', 
        'Location', 'Status', 
        # 'Opening date', 'Type', 
        'Manufacturer', 
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
    #    'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #    'height_value', 'height_unit', 
    'height_ft',  'Inversions_clean', 'Gforce_clean']]
display(df)
```

```python
df.dtypes
```

```python
# Change format
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
df.head(5)
```

```python
# Rename columns
df.columns
```

```python
df = df.rename(columns={'coaster_name': 'Coaster_Name',
                        'year_introduced': 'Year_Introduced',
                        'opening_date_clean': 'Opening_Date',
                        'speed_mph': 'Speed_mph',
                        'height_ft': 'Height_ft', 
                        'Inversions_clean': 'Inversions', 
                        'Gforce_clean': 'Gforce'})
df.head()
```

```python
null_values = df.isna().sum()
non_null_value = df.notnull().sum()
fig, ax = plt.subplots(figsize=(15,5))
# Plot null value
ax.bar(null_values.index, null_values, label='Null', color='red')
# Plot non-null values 
ax.bar(non_null_value.index, non_null_value,bottom=null_values , label='Non null', color='blue')
# ax.set_xlabel()
# plt.title()
plt.show()
```

```python
# df.isna().sum()
non_null_value = df.notnull().sum()
plt.figure(figsize=(15,5))
plt.bar(non_null_value.keys(), non_null_value.values)
plt.show()
```

**Step 5: Answer some questions (analysis)**

What are the locations with the fastest roller coasters (minimum of 10)?

```python
df1 = df.loc[df['Location']!='Other']
df1
```

```python

# df['Location'].value_counts()

# df.query('Location != "Other" ') \
#     .groupby('Location')['Speed_mph'].agg(['mean'])

df2 = df1.groupby('Location')['Speed_mph'].agg(['mean', 'count']) \
    .query('count > 10').sort_values('mean', ascending=False)
ax = df2['mean'].plot(kind='bar', title='Average Coast Speed by Location', figsize=(10,5))
ax.set_xlabel('Average Speed')
plt.show()
```

Top 5 status having the total inversions reach maximum?

```python
df3 = df.loc[~df.Status.isna()]
df3.groupby('Status')['Inversions'].agg(['sum', 'mean']).sort_values('sum', ascending=False).head(5) \
    ['sum'].plot(kind='barh')
```

With Status = Operating, Top 8 Year_Introduced having the highest average speed, height?

```python
df_operating = df.loc[df.Status == 'Operating']
df_operating.groupby('Year_Introduced')['Speed_mph'].agg(['mean']).sort_values('mean', ascending=False).head(8) \
            ['mean'].plot(kind='barh', title="Top 8 Year_Introduced")
```
