# Task 2
Check the dataset for the following 'data cleaning' issues and resolve them:
- Empty data
- Unnecessary duplicates
- Wrong formats
- Wrong values

```python
df_market.isnull().sum()
```

With "ocean_proximity" column is categorical data; therefore, we can transform NaN to value is "Unknown"

```python
df_market.loc[df_market["ocean_proximity"].isnull() == True]
```

Fill nan by Unknown

```python
df_market["ocean_proximity"] = df_market["ocean_proximity"].fillna("UNKNOWN")
```

```python
display(df_market)
df_market.isnull().sum()
```

With "median_income" is numerical data; therefore, it should be replaced by mean value.

```python
median_income_non_null = df_market.loc[~df_market.median_income.isnull(), "median_income"]
median_income_non_null
```

```python
mean_median_income = round(median_income_non_null.mean(), 4)

print(mean_median_income)
```

Fill "median_income" by mean value

```python
df_market["median_income"] = df_market["median_income"].fillna(mean_median_income).astype("Float64")
df_market
```

```python
df_market.nunique()
```

Remove duplicated records

```python
df_market.drop_duplicates()
```

After remove duplicated records, we can see that ocean_proximity column has wrong values

```python
df_market["ocean_proximity"].unique()
```

We can see that "NEAR OCEAN" is the same with "NEAR O" value, this value may be a typo error.

```python
df_market["ocean_proximity"] = df_market["ocean_proximity"].replace(["NEAR O"], "NEAR OCEAN")
df_market["ocean_proximity"].unique()
```

After remove wrong values, we can see the "ocean_proximity" column has wrong format. We will correct to right format. It should only capitalize the first character in word.

```python
df_market["ocean_proximity"] = df_market["ocean_proximity"].replace(["NEAR BAY"], "Near Bay").replace(["INLAND"], "Inland").replace(["UNKNOWN"], "Uknown").replace(["NEAR OCEAN"], "Near Ocean").replace(["ISLAND"], "Island")
df_market["ocean_proximity"].unique()
```

```python
df_market.to_csv("Pre_Processed_Market.csv")
```

# Task 3
The house prices are at the focus of this dataset. By using the new data frame, provide
the following values which describe the column 'median_house_value':
- Mean (average)
- Median
- Range, this is the difference between the maximum and the minimum values

```python
df_housing = pd.read_csv("housing.csv")
df_housing
```

```python
df_housing.isnull().sum()
```

There are 207 records which have null value on "total_bedrooms" column. Therefore, it needs to resolve null value with the suitable values. In this case, it has data type is int64; therefore, it should be replaced by median value.

```python
total_bedrooms_non_null = df_housing.loc[~df_housing.total_bedrooms.isnull(), "total_bedrooms"]
total_bedrooms_non_null
```

```python
median_total_bedrooms = round(total_bedrooms_non_null.median(), 1)

print(median_total_bedrooms)
```

```python
df_housing["total_bedrooms"] = df_housing["total_bedrooms"].fillna(median_total_bedrooms).astype("Int64")
df_housing.isnull().sum()
```

```python
df_housing.nunique()
```

Remove duplicated records if exists

```python
df_housing.drop_duplicates()
```

```python
median_house_value_non_null = df_housing.loc[~df_housing.median_house_value.isnull(), "median_house_value"]
median_house_value_non_null
```

Calculate mean of median_house_value

```python
mean_median_house_value = round(median_house_value_non_null.mean(), 1)

print(mean_median_house_value)
```

Calculate median of median_house_value

```python
median_median_house_value = round(median_house_value_non_null.median(), 1)

print(median_median_house_value)
```

To calculate Range, the difference between the maximum and the minimum values of "median_house_value"; therefore, it requires to determine max and min value of "median_house_value"

```python
max_median_house_value = median_house_value_non_null.max()
min_median_house_value = median_house_value_non_null.min()

print(f"{min_median_house_value=}")
print(f"{max_median_house_value=}")
```

```python
range_median_house_value = max_median_house_value - min_median_house_value

print(f"{range_median_house_value=}")
```

Based on the charts presented above, it was determined that the box plot chart is the best option for visualization. Because it displays the distribution of'median_house_value' numerical data. It shows the range of the central 50% of the data, with a central line representing the median value. Lines extend from each box to represent the range of the remaining data, with dots added beyond the line edges to highlight outliers.

```python
import matplotlib.pyplot as plt

# Convert median_house_value to a NumPy array
median_house_value_array = median_house_value_non_null.to_numpy()
median_house_value_index = np.arange(len(median_house_value_non_null))

# Line Chart
plt.figure(figsize=(10, 8))
plt.plot(median_house_value_index, median_house_value_array, marker = 'o', label = 'Median House Value', color = "r")
plt.xlabel("Index",  fontweight ='bold', fontsize = 15)
plt.ylabel("Median House Value (USD)", fontweight ='bold', fontsize = 15)
plt.title("Line Chart of Median House Value", fontweight ='bold', fontsize = 15)
plt.legend()
plt.show()
```

```python
# Bar Chart
barWidth = 0.75
fig = plt.subplots(figsize =(10, 8)) 
# Make the plot
plt.bar(median_house_value_index, median_house_value_array, color ='r', width = barWidth, 
            edgecolor ='grey', label ='Median House Value') 

plt.title("Bar Chart of Median House Value",  fontweight ='bold', fontsize = 15)
plt.xlabel('Index', fontweight ='bold', fontsize = 15) 
plt.ylabel('Median House Value (USD)', fontweight ='bold', fontsize = 15) 

plt.legend()
plt.show()
```

```python
# Box Plot
data = np.array([median_house_value_array], dtype=float)
T = np.transpose(data)
plt.boxplot(T)
plt.ylabel("Index", fontweight ='bold', fontsize = 10)
plt.title("Bar Chart of Median House Value", fontweight ='bold', fontsize = 10)
plt.show()
```
