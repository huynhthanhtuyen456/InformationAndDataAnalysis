# Task 5
Implement at least one of the Tasks 2 to 4 without using functions from the NumPy library, apart from initialising the array.

```python
def calculateSumArray(array):
    sum = 0
    for i in range(array.size):
        sum += array[i]
    return sum


def calculateAverageArray(array):
    sum = calculateSumArray(array=array)
    return sum / array.size


# Function to find the largest element
def largest(arr,  n, max_index=0):
    # Last index return the element
    if (max_index == n - 1):
        return arr[max_index]
 
    # Find the maximum from rest of the array
    recMax = largest(arr, n, max_index + 1)

    if recMax > arr[max_index]:
        return recMax

    return arr[max_index]


# Function to find the smallest element
def smallest(arr,  n, min_index=0):
    # Last index return the element
    if (min_index == n - 1):
        return arr[min_index]
 
    # Find the minimum from rest of the array
    recMax = largest(arr, n, min_index + 1)

    if recMax < arr[min_index]:
        return recMax

    return arr[min_index]


def sort(array: list[int], n: int):
    """Sort array in ascending order
    Parameters
    ----------
    array : [int, ...]
        An array of number such as int.

    n: int
        Length of array.

    Returns
    -------
    list[int]
        Sorted array.
    """
    for i in range(n):
        for j in range(n - i - 1):
            if (array[j] > array[j + 1]):
                temp = array[j]
                array[j] = array[j + 1]
                array[j + 1] = temp
    return array


def reverse(array, n):
    """Reverse array in ascending order
    Parameters
    ----------
    array : [int, ...]
        An array of number such as int.

    n: int
        Length of array.

    Returns
    -------
    list[int]
        Reversed array.
    """
    for i in range(n):
        for j in range(n - i - 1):
            if (array[j] < array[j + 1]):
                temp = array[j]
                array[j] = array[j + 1]
                array[j + 1] = temp
    return array


testArray = np.random.randint(0, 100, NumberOfElements, dtype=int)
print(f"Sum of {testArray=} is {calculateSumArray(testArray)}")
print(f"Average of {testArray=} is {calculateAverageArray(testArray)}")
print(f"Maximum of {testArray=} is {largest(testArray, testArray.size)}")
print(f"Minimum of {testArray=} is {smallest(testArray, testArray.size)}")
print(f"Sort array {testArray=} in ascending order into {sort(testArray, testArray.size)} ")
print(f"Reverse array {testArray=} in descending order into {reverse(testArray, testArray.size)} ")
```

# Task 6
Visualise arrays A, B and C as a table. Display the indices of the arrays in the first column of the table.

```python
def printArray(array):
    print("-" * (len("-" * 5 + " Index " + "-" * 5 + " Value" + "-" * 5) + 1))
    print("-" * 5 + " Index " + "-" * 5 + " Value " + "-" * 5)
    for i in range(len(array)):
        print("-" * 5 + f" {i} {' ' * (len('index') - len(str(i)))}" + "-" * 5 + f" {array[i]} {' ' * (len('value') - len(str(array[i])))}" + "-" * 5)
    print("-" * (len("-" * 5 + " Index " + "-" * 5 + " Value" + "-" * 5) + 1))

print("Index and value of array A in table ")
printArray(A)
print()
print("Index and value of array B in table ")
printArray(B)
print()
print("Index and value of array C in table ")
printArray(C)
print()
```
