```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_data['species'] = iris.target
display(iris_data)
```

```python
iris_data.info()
```

```python
iris_data.nunique()
```

```python
# Calculate the correlation matrix
correlation_matrix = iris_data.corr()

# Create and customize the heat map
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='viridis', linewidths=0.5, linecolor='white')
plt.title('Heat Map of Iris Dataset Correlation Matrix', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Create scatter plot with color based on iris target values
plt.scatter(iris_data['petal length (cm)'], iris_data['petal width (cm)'], c=iris.target, cmap='viridis')
plt.colorbar(label='Target Class')  # Adds a color bar to indicate target class
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Petal Length vs. Petal Width')
plt.show()
```
