# Task 1
Initialise an array containing 20 random values from the interval [0 .. 100]. Visualise the
values of the data points and plot the dataset by using a Line Chart.
Libraries:
- import numpy as np
- import pandas as pd
- import matplotlib.pyplot as plt

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats as st
```

```python
np.random.seed(42)
mu, sigma = 0, 0.1 # mean and standard deviation
sample_data = np.random.normal(mu, sigma, 20)
sample_data_index = np.arange(len(sample_data))
```

```python
# Line Chart
plt.figure(figsize=(10, 8))
plt.plot(sample_data_index, sample_data, marker = 'o', label = 'Sample Data Line Chart', color = "r")
plt.xlabel("Index",  fontweight ='bold', fontsize = 15)
plt.ylabel("Random Data", fontweight ='bold', fontsize = 15)
plt.title("Line Chart of Random Data", fontweight ='bold', fontsize = 15)
plt.legend()
plt.show()
```

# Task 2 By using functions from the NumPy library, find and print the following statistical measures:
- Mean value
- Median value
- Range
- Standard deviation

```python
# Descriptive statistics
ave_sample_data = np.mean(sample_data)
med_sample_data = np.median(sample_data)
mode_sample_data = st.mode(sample_data)
range_sample_data = np.max(sample_data) - np.min(sample_data)
std_sample_data = np.std(sample_data)

print(f"The mean of study hours (200 students) is {ave_sample_data:.5f}.\nThe median is {med_sample_data:.5f}.\nThe range is {range_sample_data:.5f}.\nThe std is {std_sample_data:5f}")
print(f"The mode of sample data: {mode_sample_data[0]:.5f}. Count {mode_sample_data[1]}")
```

# Task 5
Calculate the Interquartile range of the dataset.

```python
# Step 4: Computing the IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

print(f"{IQR=}")

#Step 5: Applying the IQR Filter
outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))

# # Step 6: Handling the Outliers
clean_data = df[~outliers.any(axis=1)]

# # Step 7: Conclusion
print(clean_data.head())
```
