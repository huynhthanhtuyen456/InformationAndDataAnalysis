```python
import networkx as nx

import matplotlib.pyplot as plt
G = nx.Graph()
```

```python
# From this point we can start adding nodes and connecting them with edges
G.add_node('Hyde Park Corner')
G.add_node('Green Park')
G.add_node('Piccadilly Circus')
G.add_node('Leicester Square')
G.add_node('Covent Garden')
G.add_node('Holborn')
G.add_edge('Hyde Park Corner', 'Green Park', weight=5)
G.add_edge('Green Park', 'Piccadilly Circus', weight=5)
G.add_edge('Piccadilly Circus', 'Leicester Square', weight=5)
G.add_edge('Leicester Square', 'Covent Garden', weight=5)
G.add_edge('Covent Garden', 'Holborn', weight=5)
```

```python
weights = nx.get_edge_attributes(G, "weight")
pos = nx.spring_layout(G)

nx.draw_networkx(G, pos=pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels=weights)

path = nx.shortest_path(G, source='Hyde Park Corner', target='Holborn')
path_edges = list(zip(path,path[1:]))
nx.draw_networkx_nodes(G,pos,nodelist=path,node_color='r')
nx.draw_networkx_edges(G,pos,edgelist=path_edges,edge_color='r',width=10)
plt.axis('equal')
plt.show()
```
