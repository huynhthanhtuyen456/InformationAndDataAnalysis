```python
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats as st
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
study_hours = np.random.normal(mu, sigma, 200)
```

```python
ave_hours = np.mean(study_hours)
med_hours = np.median(study_hours)
mode_hours = st.mode(study_hours)
```

```python
# visualize data
plt.figure()
plt.hist(study_hours, bins=20, color="skyblue", edgecolor="black")
plt.title("Study Hours of 200 students")
plt.xlabel("Hours studied per week")
plt.ylabel("Frequency")
plt.axvline(ave_hours, color="red", linestyle="dashed", linewidth=1.2, label=f"Mean:{ave_hours:.5f}")
plt.axvline(med_hours, color="green", linewidth=1.2, label=f"Median:{med_hours:.5f}")
plt.legend()
plt.show()
```

# Interential Statistics

Hypothesis Testing
 - Null Hypothesis $H_0: \mu = 14$
 - Alternative Hypothesis $H_1: \mu \neq 14$

```python
# Confidence Interval: 95%
confidence_level = 0.95
degree_of_freedom = int(len(study_hours) - 1)
confidence_interval = st.t.interval(confidence_level, degree_of_freedom)
confidence_interval

print(f"{confidence_interval[0]:.5f}")
print(f"{confidence_interval[1]:.5f}")
```

```python
mean_test = 14
t_stats, p_value = st.ttest_1samp(study_hours, mean_test)
print(f"Hypothesis testing (H_0): mean = {mean_test} (H_1): mean != {mean_test}")
print(f"T statistics: {t_stats:.5f}")
print(f"P value: {p_value:.5f}")

# Reject or not Null Hypothesis
if p_value < 0.05:
    print(f"We reject the Null Hypothesis: The average study hours is significantlt different from {mean_test}")
else:
    print(f"We fail to reject the Null Hypothesis: No different to {mean_test}")
```
