```python
df_sample = pd.read_csv("sample1.csv", delimiter=";")
df_sample
```

```python
df_sample.info()
```

```python
df_sample["Quantity"]= pd.to_numeric(df_sample["Quantity"], errors="coerce")
df_sample["Price"] = pd.to_numeric(df_sample["Price"], errors="coerce")

quantity_non_null = df_sample.loc[~df_sample.Quantity.isnull(), "Quantity"]
quantity_non_null
mean_quantity = round(quantity_non_null.mean(),0)

print(mean_quantity)

df_sample["Quantity"] = df_sample["Quantity"].fillna(mean_quantity).astype(int)
df_sample
```

```python
import matplotlib.pyplot as plt

df_sample.Quantity.hist()
plt.show()
```
