```python
import matplotlib.pyplot as plt
```

```python
# Define data
categories = ['A', 'B', 'C']
values = [10, 15, 7]

# Define colors using RGB model
colors = ['#FF5733', '#33FF57', '#3357FF']

# Create bar chart
plt.bar(categories, values, color=colors)

# Set font color
plt.xlabel('Category', color='#000000')
plt.ylabel('Values', color='#000000')
plt.title('RGB Color Model Example', color='#000000')

# Show plot
plt.show()
```

```python
# Define data
categories = ['A', 'B', 'C']
values = [10, 15, 7]

# Define RGB values for colors
colors_rgb = [(20, 110, 60), (50, 170, 200), (200, 50, 150)]

# Convert RGB values to hexadecimal format
colors_hex = ['#{:02x}{:02x}{:02x}'.format(r, g, b) for r, g, b in colors_rgb]

# Create bar chart with dynamic colors
plt.bar(categories, values, color=colors_hex)

# Set font color
plt.xlabel('Category', color='#000000')
plt.ylabel('Values', color='#000000')
plt.title('Dynamic RGB Colors Example', color='#000000')

# Show plot
plt.show()
```
